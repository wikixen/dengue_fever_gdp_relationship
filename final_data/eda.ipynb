```python
import pandas as pd
import numpy as numpy
pd.set_option('display.max_columns', 500)
pd.set_option('display.max_rows', 500)
df = pd.read_csv('merge.csv')
```

```python
df.head(3)
```

```python
df2 = df.pivot_table(['gdp_val','total_cases','total_deaths'],['entity'],'year').reset_index()
df2.columns = [ '_'.join([str(c) for c in c_list]) for c_list in df2.columns.values ]
df2.head()
```

# Tables
- GDP
- GDP growth
- Cases
- Deaths

### GDP & Growth

```python
#First col is 0 because theres nothing to comapre it to
gdp = df2.iloc[:,:21]
gdp_growth = gdp.iloc[:,1:21].pct_change(periods=1, axis=1).round(2)*100
gdp_growth2 = gdp[['entity_']].join(gdp_growth)
gdp_growth2.fillna(0, inplace=True)
gdp_growth2.head()
```

#### I needed the below for tableau.

```python
gdp_growth3 = pd.melt(gdp_growth2, id_vars=['entity_'],var_name=['year']).sort_values(by='entity_').reset_index()
gdp_growth3['year'] = gdp_growth3['year'].str[8:]
gdp_growth3.drop(columns=['index'], inplace= True)
gdp_growth3
```

```python
gdp_growth3.to_csv('gdp_growth2.csv')
```

### Cases & Deaths

#### Cases

```python
cases = df2[['entity_']].join(df2.iloc[:,21:41])
cases.head(3)
```

```python
cases[['entity_']].join(cases.iloc[:,1:].agg(['mean','min','median','max'],axis=1)).sort_values(by='mean', ascending=False).reset_index()
```

#### Deaths

```python
deaths = df2[['entity_']].join(df2.iloc[:,41:])
deaths.head(3)
```

```python
deaths[['entity_']].join(deaths.iloc[:,1:].agg(['mean','min','median','max'],axis=1)).sort_values(by='mean', ascending=False).reset_index()
```
